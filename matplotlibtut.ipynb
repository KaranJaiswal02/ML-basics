# Matplot lib
## Scatterplot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
rollno = [1,2,3,4,5,6,7,8,9]
marks = [10,20,30,40,50,60,70,80,90]
plt.style.use('dark_background')
#the above statement is used for the change of the background 
#color of the map/graph drawn
```

```python
plt.scatter(rollno, marks)
plt.show()
#use color = 'green ' and change the color of the dots
```

```python
plt.scatter(rollno, marks , color = 'green', marker = '*')
plt.show()
#marker changes the shape of the scatter plot
```

```python
# more fuctions using the plot function
```

```python
plt.figure(figsize = (4,4))
plt.plot(rollno , marks , 'bd', markersize = 10)
plt.show()
```
